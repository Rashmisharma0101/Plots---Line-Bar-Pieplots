```python
#Scatter plot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x= [1,2,3,4,5,6]
y= [10,8,7,4,6,6]
colors = [1,2,3,4,5,6]
plt.scatter (x,y, edgecolor = 'black', s=100, alpha = 0.75, c=colors, cmap = 'Greens')
cbar = plt.colorbar()
cbar.set_label('Satisfaction level')
plt.grid()
plt.show()
```

```python
#histograms
ages = [10,20,45,55,23,34]
bins = [10,20,30,40,50,60]
median_age = 44.5
plt.hist(ages, bins=bins, edgecolor= 'red', color ='Grey', alpha = 0.75)
plt.axvline(median_age, color= 'black', label = 'Median')  #axisvertical line -axvline
plt.legend()
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.title('People belonging to different age groups')
plt.show()
```

```python
#Stack Plot,time contributed by each person in a project per day
minutes = [1,2,3,4,5,6]
p1 = [18,23,32,44,67,91]
p2 = [11, 19, 32, 61, 80, 94]
plt.stackplot(minutes, p1, p2, alpha = 0.45, edgecolor = 'grey')
plt.legend(['person1', 'person2'], loc= 'upper left')
plt.xlabel('Days')
plt.ylabel('Hours Contributed')
plt.title('People working in project')
plt.tight_layout()
plt.show()
```
